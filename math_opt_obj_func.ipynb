### Task 2: Mathematical optimization of an objective function


```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Defining the objective function
def objective_function(xy):
    x ,y = xy
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

# Defining the range of validity of x and y
range_of_validity = (-5, 5)

# Using the optimization algorithm to minimize the objective function
result = minimize(objective_function, (0, 0), method='Nelder-Mead')

# Printing the optimized results
print("Optimized Results: ")
print(result)

# Creating a 2D visualization of the objective function in x, y - space
x = np.linspace(range_of_validity[0], range_of_validity[1], 100)
y = np.linspace(range_of_validity[0], range_of_validity[1], 100)
X, Y = np.meshgrid(x, y)
Z = objective_function([X, Y])

fig, ax = plt.subplots()
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
ax.plot(result.x[0], result.x[1], 'ro', label='Optimized Point')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Objective Function Visualization')
plt.show()
```
